```python
import pandas as pd
import matplotlib.pyplot as plt

datafile = '../data/country_gdp_growth.csv'
year = '2009'
cols = ['Country Name', 'Country Code', year]
df = pd.read_csv(datafile, skiprows=4, usecols=cols)

df = df.dropna()
df = df.sort_values(by=[year], ascending=True)
df['positive'] = df[year] > 0
```

```python
f, ax = plt.subplots(figsize=(45, 20))
ax.bar(df['Country Code'], df[year], color=df.positive.map({True: 'g', False: 'r'}))
ax.tick_params(axis='x', rotation=90)
ax.set_title('BDP rast drzav leta {}'.format(year), fontdict={'fontsize': 36})
```

```python
top_economies = ['USA', 'CHN', 'JPN', 'DEU', 'GIB', 'IND', 'FRA', 'BRA', 'ITA', 'CAN']
df = df[df['Country Code'].isin(top_economies)]

f, ax = plt.subplots()
ax.bar(df['Country Name'], df[year], color=df.positive.map({True: 'g', False: 'r'}))
ax.tick_params(axis='x', rotation=90)
ax.set_title('BDP rast 10 držav z največjimi ekonomijami leta {}'.format(year))
```

```python
plt.gcf().subplots_adjust(bottom=2)
ax.get_figure().savefig('../img/top10_country_growth_gdp_{}.png'.format(year))
```
