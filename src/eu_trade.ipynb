```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

datafile = '../data/country_exports.csv'
year = '2008'
next_year = str(int(year) + 1)
cols = ['Country Name', 'Country Code', year, next_year]
df = pd.read_csv(datafile, skiprows=4, usecols=cols)

df = df.dropna()
df = df.sort_values(by=[year, next_year], ascending=True)
df.head()
```

```python
eu = ['AUT', 'BEL', 'BGR', 'HRV', 'CYP', 'CZE', 'DNK', 'EST', 'FIN', 'FRA', 'DEU', 'GRC', 'HUN', 'IRL', 'ITA', 'LVA', 'LTU',
     'LUX', 'MLT', 'NLD', 'POL', 'PRT', 'ROU', 'SVK', 'SVN', 'ESP', 'SWE', 'GBR']
df = df[df['Country Code'].isin(eu)]

f, ax = plt.subplots(figsize=(16, 12))
ax.bar(df['Country Name'], df[year], color='green')
ax.bar(df['Country Name'], df[next_year], color='palegreen')
ax.tick_params(axis='x', rotation=90)
ax.set_title('Vrednost izvozov ƒçlanic EU ({}-{})'.format(year, next_year))
ax.set_ylabel('Vrednost izvozov v bilijonih')

green_patch = mpatches.Patch(color='green', label='Vrednost izvoza leta {}'.format(year))
pgreen_patch = mpatches.Patch(color='palegreen', label='Vrednost izvoza leta {}'.format(next_year))
ax.legend(handles=[green_patch, pgreen_patch])
ax.get_legend().set_title('Vrednost izvozov')
ax.yaxis.get_offset_text().set_visible(False)
```
