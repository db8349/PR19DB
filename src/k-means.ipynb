```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import numpy as np

datafile = '../data/country_gdp_growth.csv'
year = '2008'
next_year = str(int(year) + 1)
cols = ['Country Name', 'Country Code', year, next_year]
df = pd.read_csv(datafile, skiprows=4, usecols=cols)

df = df.dropna()
df
```

```python
X = np.array(df[[year, next_year]])
kmeans = KMeans(n_clusters=2)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)

f, ax = plt.subplots(figsize=(16, 13))

ax.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=30, cmap='RdYlGn')
ax.set_title('KMeans gručenje držav po rasti BDP-ja ({}-{})'.format(year, next_year))
ax.set_xlabel('Rast leta {}'.format(year))
ax.set_ylabel('Rast leta {}'.format(next_year))

centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='yellow', s=500, alpha=0.7)
```
