```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

datafile = '../data/country_investments.csv'
year = '2016'
next_year = str(int(year) + 1)
cols = ['Country Name', 'Country Code', year, next_year]
df = pd.read_csv(datafile, skiprows=4, usecols=cols)

df = df.dropna()
df = df.sort_values(by=[year, next_year], ascending=True)
df.head()
```

```python
eu = ['AUT', 'BEL', 'BGR', 'HRV', 'CYP', 'CZE', 'DNK', 'EST', 'FIN', 'FRA', 'DEU', 'GRC', 'HUN', 'IRL', 'ITA', 'LVA', 'LTU',
     'LUX', 'MLT', 'NLD', 'POL', 'PRT', 'ROU', 'SVK', 'SVN', 'ESP', 'SWE', 'GBR']
df = df[df['Country Code'].isin(eu)]
df
```

```python
f, ax = plt.subplots(figsize=(16, 12))
ax.bar(df['Country Name'], df[next_year], color='paleturquoise')
ax.bar(df['Country Name'], df[year], color='blue')
ax.tick_params(axis='x', rotation=90)
ax.set_title('Vrednost tujih investicij v članicah EU ({}-{})'.format(year, next_year))
ax.set_ylabel('Vrednost investicij v 100 milijardah')

blue_patch = mpatches.Patch(color='blue', label='Vrednost tujih investicij leta {}'.format(year))
pblue_patch = mpatches.Patch(color='paleturquoise', label='Vrednost tujih investicij leta {}'.format(next_year))
ax.legend(handles=[blue_patch, pblue_patch])
ax.get_legend().set_title('Vrednost investicij')
ax.yaxis.get_offset_text().set_visible(False)
```

```python
ax.get_figure().savefig('../img/eu_investments_({}-{}).png'.format(year, next_year), bbox_inches = "tight")
```
